# OOP Recommendation System using Genetic Algorithm

```python
import random
import itertools
import numpy as np
import pandas as pd

class BookRecommendationSystem:
    def __init__(self, user_ratings_input, user_ratings, semantic):
        self.user_ratings_input = user_ratings_input
        self.user_ratings = user_ratings
        self.semantic = semantic

    def calculate_semrating(self, solution):
        semrating = []
        for i, book_i in enumerate(solution):
            for j in range(i + 1, len(solution)):
                f11 = sum(1 for value_i, value_j in zip(self.semantic[book_i], self.semantic[solution[j]]) if value_i == 1 and value_j == 1)
                f10 = sum(1 for value_i, value_j in zip(self.semantic[book_i], self.semantic[solution[j]]) if value_i == 1 and value_j == 0)
                f01 = sum(1 for value_i, value_j in zip(self.semantic[book_i], self.semantic[solution[j]]) if value_i == 0 and value_j == 1)
                rating = f11 / (f10 + f01 + f11)
                semrating.append(rating)
        return semrating

    def possible_solutions(self):
        missing_items = [index for index, rating in enumerate(self.user_ratings_input) if rating == 0]
        all_recommendations = []
        for r in range(2, len(missing_items) + 1):  # Modified: Start from 2 instead of 1
            combinations = itertools.combinations(missing_items, r)
            all_recommendations.extend(list(combinations))
        return all_recommendations

    def generate_initial_population(self, all_recommendations):
        population = []
        fitness_scores = [self.calculate_total_rating(solution) for solution in all_recommendations]
        fitness_sum = sum(fitness_scores)
        for _ in range(len(all_recommendations)):
            selected_solution = None
            while selected_solution is None:
                random_number = random.random()
                for i, fitness_score in enumerate(fitness_scores):
                    cumulative_probability = sum(fitness_scores[:i+1]) / fitness_sum
                    if random_number <= cumulative_probability:
                        selected_solution = all_recommendations[i]
                        break
            population.append(selected_solution)
        return population

    def calculate_total_rating(self, solution):
        total_predicted_rating = 0
        for item_index in solution:
            predicted_rating = self.calculate_predicted_rating_aws(item_index)
            total_predicted_rating += predicted_rating
        return total_predicted_rating

    def calculate_predicted_rating_aws(self, target_item_index):
        average_rating_target_user = self.average_rating(self.user_ratings_input)
        total_weighted_rating = 0
        total_similarity = 0
        for user_rating in self.user_ratings:
            if user_rating[target_item_index] != 0:
                similarity = self.cosine_similarity(self.user_ratings_input, user_rating)
                average_rating_user_j = self.average_rating(user_rating)
                adjusted_rating = user_rating[target_item_index] - average_rating_user_j
                total_weighted_rating += similarity * adjusted_rating
                total_similarity += abs(similarity)
        if total_similarity != 0:
            predicted_rating = average_rating_target_user + (total_weighted_rating / total_similarity)
        else:
            predicted_rating = average_rating_target_user
        return predicted_rating

    def cosine_similarity(self, a, b):
        return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

    def average_rating(self, user_rating):
        if isinstance(user_rating, int):
            return user_rating
        ratings = [rating for rating in user_rating if rating != 0]
        return np.mean(ratings)

    def selection(self, population, fitness_scores):
        selected_population = []
        total_fitness = sum(fitness_scores)
        if total_fitness == 0:
            probabilities = [1 / len(population)] * len(population)
        else:
            probabilities = [fitness_score / total_fitness for fitness_score in fitness_scores]
        for _ in range(len(population)):
            selected_solution = random.choices(population, probabilities)[0]
            selected_population.append(selected_solution)
        return selected_population

    def one_point_crossover(self, parent1, parent2):
        parent1 = list(parent1)
        parent2 = list(parent2)
        crossover_point = random.randint(0, len(parent1))  # Modified: Allow crossover at the beginning or end
        child1 = parent1[:crossover_point] + parent2[crossover_point:]
        child2 = parent2[:crossover_point] + parent1[crossover_point:]
        child1 = self.remove_duplicates(child1)
        child2 = self.remove_duplicates(child2)
        return tuple(child1), tuple(child2)

    def uniform_crossover(self, parent1, parent2):
        parent1 = list(parent1)
        parent2 = list(parent2)
        child1 = []
        child2 = []
        for gene1, gene2 in zip(parent1, parent2):
            if random.random() < 0.5:
                child1.append(gene1)
                child2.append(gene2)
            else:
                child1.append(gene2)
                child2.append(gene1)
        child1 = self.remove_duplicates(child1)
        child2 = self.remove_duplicates(child2)
        return tuple(child1), tuple(child2)

    def swap_mutation(self, solution, all_recommendations, mutation_rate):
        mutated_solution = list(solution)
        for i in range(len(solution)):
            if random.random() < mutation_rate:
                possible_items = all_recommendations[i % len(all_recommendations)]
                mutated_solution[i] = random.choice(possible_items)
        mutated_solution = self.remove_duplicates(mutated_solution)
        return tuple(mutated_solution)


    def inversion_mutation(self, solution):
        mutated_solution = list(solution)
        if len(mutated_solution) <= 1:
            return mutated_solution
        start = random.randint(0, len(mutated_solution) - 1)
        end = random.randint(start + 1, len(mutated_solution))
        mutated_solution[start:end] = reversed(mutated_solution[start:end])
        return mutated_solution

    @staticmethod
    def remove_duplicates(solution):
        seen = set()
        return [x for x in solution if not (x in seen or seen.add(x))]

    def genetic_algorithm(self, mutation_rate, num_generations, crossover_func, mutation_func):
        all_recommendations = self.possible_solutions()
        semratings = [self.calculate_semrating(solution) for solution in all_recommendations]
        population = self.generate_initial_population(all_recommendations)

        best_solution = None
        best_total_rating = -float('inf')
        best_semrating = -float('inf')

        for generation in range(num_generations):
            fitness_scores = [self.calculate_total_rating(solution) for solution in population]
            semratings = [sum(self.calculate_semrating(solution)) for solution in population]

            best_solution_index = np.argmax(fitness_scores)
            current_total_rating = fitness_scores[best_solution_index]
            current_semrating = semratings[best_solution_index]

            if current_total_rating > best_total_rating or (current_total_rating == best_total_rating and current_semrating > best_semrating):
                best_solution = population[best_solution_index]
                best_total_rating = current_total_rating
                best_semrating = current_semrating

            selected_population = self.selection(population, fitness_scores)
            next_population = []

            for _ in range(len(population) // 2):
                parent1, parent2 = random.sample(selected_population, 2)

                if crossover_func == "one_point":
                    child1, child2 = self.one_point_crossover(parent1, parent2)
                elif crossover_func == "uniform":
                    child1, child2 = self.uniform_crossover(parent1, parent2)
                else:
                    raise ValueError("Invalid crossover function: " + crossover_func)

                if mutation_func == "swap":
                    mutated_child1 = self.swap_mutation(child1, all_recommendations,mutation_rate)
                    mutated_child2 = self.swap_mutation(child2, all_recommendations,mutation_rate)
                elif mutation_func == "inversion":
                    mutated_child1 = self.inversion_mutation(child1)
                    mutated_child2 = self.inversion_mutation(child2)
                else:
                    raise ValueError("Invalid mutation function: " + mutation_func)

                next_population.extend([mutated_child1, mutated_child2])

            population = next_population
            population.append(best_solution)

        best_solution = sorted(best_solution)
        best_solution = [item for item in best_solution if self.calculate_predicted_rating_aws(item) > 3]
        return best_solution


    def calculate_mae(self, user_ratings):
        total_mae = 0
        num_users = len(user_ratings)
        num_items = len(user_ratings[0])

        for user_index, user_rating in enumerate(user_ratings):
            filtered_user_ratings = user_ratings[:user_index] + user_ratings[user_index + 1:]

            error = 0
            count = 0

            for index in range(num_items):
                if user_rating[index] != 0:
                    current_item_predicted_rating = self.calculate_predicted_rating_aws(index)
                    item_mae = abs(current_item_predicted_rating - user_rating[index])
                    error += item_mae
                    count += 1

            if count > 0:
                mae = error / count
                total_mae += mae

        if num_users > 0 and num_items > 0:
            total_mae = total_mae / (num_users * num_items)

        return total_mae

    def calculate_precision(self, actual_ratings, predicted_ratings, threshold=3):
        true_positives = sum(1 for actual, predicted in zip(actual_ratings, predicted_ratings) if actual >= threshold and predicted >= threshold)
        predicted_positives = sum(1 for predicted in predicted_ratings if predicted >= threshold)
        precision = true_positives / predicted_positives if predicted_positives > 0 else 0
        return precision

    def calculate_recall(self, actual_ratings, predicted_ratings, threshold=3):
        true_positives = sum(1 for actual, predicted in zip(actual_ratings, predicted_ratings) if actual >= threshold and predicted >= threshold)
        actual_positives = sum(1 for actual in actual_ratings if actual >= threshold)
        recall = true_positives / actual_positives if actual_positives > 0 else 0
        return recall

    def calculate_f1_measure(self, actual_ratings, predicted_ratings, threshold=3):
        precision = self.calculate_precision(actual_ratings, predicted_ratings, threshold)
        recall = self.calculate_recall(actual_ratings, predicted_ratings, threshold)
        f1_measure = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
        return f1_measure

    def evaluation(self, best_solution):
        actual_ratings = self.user_ratings_input
        predicted_ratings = [self.calculate_predicted_rating_aws(item) for item in best_solution]
        mae = self.calculate_mae(self.user_ratings)
        precision = self.calculate_precision(actual_ratings, predicted_ratings, threshold=3)
        recall = self.calculate_recall(actual_ratings, predicted_ratings, threshold=3)
        f1_measure = self.calculate_f1_measure(actual_ratings, predicted_ratings, threshold=3)
        return mae, precision, recall, f1_measure

    def hyperparameter_tuning(self, population_size_range, mutation_rate_range, num_generations_range, crossover_funcs, mutation_funcs):
        results = []
        for population_size in population_size_range:
            for mutation_rate in mutation_rate_range:
                for num_generations in num_generations_range:
                    for crossover_func in crossover_funcs:
                        for mutation_func in mutation_funcs:
                            start_time = time.time()
                            best_solution = self.genetic_algorithm(mutation_rate, num_generations, crossover_func, mutation_func)
                            elapsed_time = time.time() - start_time
                            best_total_rating = self.calculate_total_rating(best_solution)
                            best_semrating = sum(self.calculate_semrating(best_solution))
                            results.append((population_size, mutation_rate, num_generations, crossover_func, mutation_func, best_total_rating, best_semrating, best_solution, elapsed_time))
        results_df = pd.DataFrame(results, columns=["population_size", "mutation_rate", "num_generations", "crossover_func", "mutation_func", "best_total_rating", "best_semrating", "best_solution", "elapsed_time"])
        results_df.sort_values(by=["elapsed_time"], ascending=True, inplace=True)
        return results_df
```

### MAE

```python
def calculate_mae(user_ratings):
    total_mae = 0
    num_users = len(user_ratings)
    num_items = len(user_ratings[0])

    for user_index, user_rating in enumerate(user_ratings):
        filtered_user_ratings = user_ratings[:]
        filtered_user_ratings.pop(user_index)

        error = 0
        count = 0

        for index in range(num_items):
            if user_rating[index] != 0:
                current_item_predicted_rating = calculate_predicted_rating_aws(user_rating, index, filtered_user_ratings)
                item_mae = abs(current_item_predicted_rating - user_rating[index])
                error += item_mae
                count += 1

        if count > 0:
            mae = error / count
            total_mae += mae

    if num_users > 0 and num_items > 0:
        total_mae = total_mae / (num_users * num_items)

    return total_mae

mae = calculate_mae(user_ratings)
print("MAE:", mae)
```

### Precision, Recall, F1-Measure

```python
def calculate_precision(actual_ratings, predicted_ratings, threshold=3):
    true_positives = sum(1 for actual, predicted in zip(actual_ratings, predicted_ratings) if actual >= threshold and predicted >= threshold)
    predicted_positives = sum(1 for predicted in predicted_ratings if predicted >= threshold)
    precision = true_positives / predicted_positives if predicted_positives > 0 else 0
    return precision

def calculate_recall(actual_ratings, predicted_ratings, threshold=3):
    true_positives = sum(1 for actual, predicted in zip(actual_ratings, predicted_ratings) if actual >= threshold and predicted >= threshold)
    actual_positives = sum(1 for actual in actual_ratings if actual >= threshold)
    recall = true_positives / actual_positives if actual_positives > 0 else 0
    return recall

def calculate_f1_measure(actual_ratings, predicted_ratings, threshold=3):
    precision = calculate_precision(actual_ratings, predicted_ratings, threshold)
    recall = calculate_recall(actual_ratings, predicted_ratings, threshold)
    f1_measure = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
    return f1_measure

# Example usage
actual_ratings = [4, 2, 3, 5, 4]
predicted_ratings = [3, 1, 2, 4, 3]

precision = calculate_precision(actual_ratings, predicted_ratings, threshold=3)
recall = calculate_recall(actual_ratings, predicted_ratings, threshold=3)
f1_measure = calculate_f1_measure(actual_ratings, predicted_ratings, threshold=3)

print("Precision:", precision)
print("Recall:", recall)
print("F1-measure:", f1_measure)
```
